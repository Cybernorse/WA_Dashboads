```python
import os
import pandas as pd
import librosa
import numpy as np
```

```python
def waveforms(dir):
    
    adata=[i for i in os.listdir(dir) if i.split('.')[1]=='webm' or i.split('.')[1]=='ogg' or i.split('.')[1]=='wav']
    bim=[]
    count=0
    for i in adata:
        if i.split('-')[2] == '01':
            emotion = 'neutral'
        elif i.split('-')[2] == '02':
            emotion = 'calm'
        elif i.split('-')[2] == '03':
            emotion = 'happy'
        elif i.split('-')[2] == '04':
            emotion = 'sad'
        elif i.split('-')[2] == '05':
            emotion = 'angry'
        elif i.split('-')[2] == '06':
            emotion = 'fear'
        elif i.split('-')[2] == '07':
            emotion = 'disgust'
        elif i.split('-')[2] == '08':
            emotion = 'surprise'
        else:
            emotion = 'unknown'
            
        data_x, sampling_rate = librosa.load(dir+i,res_type='kaiser_fast')
        mfccs = np.mean(librosa.feature.mfcc(y=data_x, sr=sampling_rate, n_mfcc=40).T,axis=0)
        bim.append([mfccs,emotion,i])
        count+=1
        print(f"************** {count} **************")
    return bim   
```

```python
bim=waveforms('/home/bigpenguin/dataset&notebook/emotions/emods/emos/')
```

```python
frq=[]
labels=[]
id=[]
for i in bim:
    for x in i:
        if len(x)==40:
            frq.append(x)
        if len(x)!=40 and len(x)!=24:
            labels.append(x)
        if len(x)==24:
            id.append(x)
```

```python
df=pd.DataFrame(frq)
df['status']=labels
df['id']=id
df
```
